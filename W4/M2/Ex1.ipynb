I used DocumentConverter() to convert a PDF from a URL into a Docling document. Then I exported the converted document to Markdown using export_to_markdown(). The output showed that Docling can extract the document structure and text into a readable Markdown format.

```python
import json
from pathlib import Path

from docling.datamodel.accelerator_options import AcceleratorDevice, AcceleratorOptions
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions, TableStructureOptions
from docling.document_converter import DocumentConverter, PdfFormatOption

# Use the same PDF URL that worked in Task 1
input_doc_path = "https://arxiv.org/pdf/2408.09869"

# Custom PDF pipeline options
pipeline_options = PdfPipelineOptions()
pipeline_options.do_ocr = True
pipeline_options.do_table_structure = True
pipeline_options.table_structure_options = TableStructureOptions(do_cell_matching=True)

# English OCR
pipeline_options.ocr_options.lang = ["en"]

pipeline_options.accelerator_options = AcceleratorOptions(
    num_threads=4,
    device=AcceleratorDevice.AUTO
)

doc_converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
    }
)

conv_result = doc_converter.convert(input_doc_path)

output_dir = Path("scratch_custom")
output_dir.mkdir(parents=True, exist_ok=True)

doc_filename = "docling_technical_report"

# Save outputs
with (output_dir / f"{doc_filename}.json").open("w", encoding="utf-8") as fp:
    fp.write(json.dumps(conv_result.document.export_to_dict()))

with (output_dir / f"{doc_filename}.txt").open("w", encoding="utf-8") as fp:
    fp.write(conv_result.document.export_to_markdown(strict_text=True))

with (output_dir / f"{doc_filename}.md").open("w", encoding="utf-8") as fp:
    fp.write(conv_result.document.export_to_markdown())

with (output_dir / f"{doc_filename}.doctags").open("w", encoding="utf-8") as fp:
    fp.write(conv_result.document.export_to_doctags())

print("Custom conversion completed.")
print("Files saved in:", output_dir)
```

I customized the PDF conversion pipeline by enabling OCR and table structure detection. This means Docling was not just extracting text, but also trying to understand document layout and tables. The results were saved as JSON, plain text, Markdown, and doctags.

```python
import json
import yaml
from pathlib import Path

from docling.datamodel.base_models import ConversionStatus
from docling.datamodel.pipeline_options import PdfPipelineOptions
from docling.document_converter import DocumentConverter
from docling_core.types.doc import ImageRefMode

# Use online PDFs instead of local tests/data files
input_doc_paths = [
    "https://arxiv.org/pdf/2408.09869",
    "https://arxiv.org/pdf/2206.01062",
    "https://arxiv.org/pdf/2203.01017",
]

output_dir = Path("scratch_batch")
output_dir.mkdir(parents=True, exist_ok=True)

pipeline_options = PdfPipelineOptions()
pipeline_options.generate_page_images = True

doc_converter = DocumentConverter()

conv_results = doc_converter.convert_all(
    input_doc_paths,
    raises_on_error=False
)

success_count = 0
partial_success_count = 0
failure_count = 0

for i, conv_res in enumerate(conv_results, start=1):
    if conv_res.status == ConversionStatus.SUCCESS:
        success_count += 1
        doc_filename = f"batch_document_{i}"

        conv_res.document.save_as_json(
            output_dir / f"{doc_filename}.json",
            image_mode=ImageRefMode.PLACEHOLDER
        )

        conv_res.document.save_as_html(
            output_dir / f"{doc_filename}.html",
            image_mode=ImageRefMode.EMBEDDED
        )

        conv_res.document.save_as_markdown(
            output_dir / f"{doc_filename}.md",
            image_mode=ImageRefMode.PLACEHOLDER
        )

        conv_res.document.save_as_markdown(
            output_dir / f"{doc_filename}.txt",
            image_mode=ImageRefMode.PLACEHOLDER,
            strict_text=True
        )

        with (output_dir / f"{doc_filename}.yaml").open("w", encoding="utf-8") as fp:
            fp.write(yaml.safe_dump(conv_res.document.export_to_dict()))

    elif conv_res.status == ConversionStatus.PARTIAL_SUCCESS:
        partial_success_count += 1
        print("Partial success:", conv_res.input.file)

    else:
        failure_count += 1
        print("Failed:", conv_res.input.file)

print("Batch conversion completed.")
print("Success:", success_count)
print("Partial success:", partial_success_count)
print("Failed:", failure_count)
print("Files saved in:", output_dir)
```

```python
import json
import yaml
from pathlib import Path

from docling.datamodel.base_models import InputFormat
from docling.document_converter import DocumentConverter

input_paths = [
    "https://arxiv.org/pdf/2408.09869",
    Path("sample.md"),
    Path("sample.html"),
]

output_dir = Path("scratch_multiformat")
output_dir.mkdir(parents=True, exist_ok=True)

doc_converter = DocumentConverter(
    allowed_formats=[
        InputFormat.PDF,
        InputFormat.HTML,
        InputFormat.MD,
    ]
)

conv_results = doc_converter.convert_all(
    input_paths,
    raises_on_error=False
)

for i, res in enumerate(conv_results, start=1):
    if res.document is None:
        print("Skipped or failed:", res.input.file)
        continue

    doc_filename = f"multiformat_document_{i}"

    print(f"Document {i} converted.")

    with (output_dir / f"{doc_filename}.md").open("w", encoding="utf-8") as fp:
        fp.write(res.document.export_to_markdown())

    with (output_dir / f"{doc_filename}.json").open("w", encoding="utf-8") as fp:
        fp.write(json.dumps(res.document.export_to_dict()))

    with (output_dir / f"{doc_filename}.yaml").open("w", encoding="utf-8") as fp:
        fp.write(yaml.safe_dump(res.document.export_to_dict()))

print("Multi-format conversion completed.")
print("Files saved in:", output_dir)
```

- FAIR
- GOOD

The main document-level confidence fields are:
- `mean_grade`: shows the average quality of the conversion
- `low_grade`: shows the lower-end quality and helps identify weak/problem areas

The component scores are:
- `layout_score`: quality of document layout recognition
- `ocr_score`: quality of OCR-extracted content
- `parse_score`: quality of digital text parsing
- `table_score`: quality of table extraction

Observation:
Confidence scores help decide whether the converted document is reliable or needs manual review/post-processing.
